```python
# ========================================
# CONFIGURACIÃ“N STORAGE CON MANAGED IDENTITY
# ========================================

storage_account = "adbdatalake01111"

spark.conf.set(
    f"fs.azure.account.auth.type.{storage_account}.dfs.core.windows.net",
    "OAuth"
)

spark.conf.set(
    f"fs.azure.account.oauth.provider.type.{storage_account}.dfs.core.windows.net",
    "org.apache.hadoop.fs.azurebfs.oauth2.MsiTokenProvider"
)
```

```python
raw_base_path = f"abfss://raw@{storage_account}.dfs.core.windows.net"

movies_df = (
    spark.read
    .option("header", "true")
    .csv(f"{raw_base_path}/movies/Movies.csv")
)

film_details_df = (
    spark.read
    .option("header", "true")
    .csv(f"{raw_base_path}/movies/FilmDetails.csv")
)

more_info_df = (
    spark.read
    .option("header", "true")
    .csv(f"{raw_base_path}/movies/MoreInfo.csv")
)

poster_path_df = (
    spark.read
    .option("header", "true")
    .csv(f"{raw_base_path}/movies/PosterPath.csv")
)
```

```python
movies_df.write.mode("overwrite").parquet(
    f"{raw_base_path}/movies_parquet/movies"
)

film_details_df.write.mode("overwrite").parquet(
    f"{raw_base_path}/movies_parquet/film_details"
)

more_info_df.write.mode("overwrite").parquet(
    f"{raw_base_path}/movies_parquet/more_info"
)

poster_path_df.write.mode("overwrite").parquet(
    f"{raw_base_path}/movies_parquet/poster_path"
)
```
